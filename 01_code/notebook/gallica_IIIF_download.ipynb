```python
import requests
import time
from tqdm import tqdm
import os
import random
```

```python
manifest = "https://gallica.bnf.fr/iiif/ark:/12148/bpt6k65671403/manifest.json"
headers= {
    "user1":{
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
'Accept': 'application/json'
        },
    "user2":{
        'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 14541.0.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36',
        'Accept' : 'application/json'
        },
    "user3":{
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:15.0) Gecko/20100101 Firefox/15.0.1',
        'Accept' : 'application/json'
        },
    
}
```

```python
resultat = requests.get(manifest, headers=headers['user1'])
resultat
data = resultat.json()
```

```python
liste_url= []
for canvas in data.get('sequences')[0].get('canvases'):
  images = canvas.get('images')
  for image in images:
    resource = image.get('resource')
    if resource.get('@id'):
      url = resource.get('@id')
      liste_url.append(url)

print(liste_url)
```

```python
n = 0
for url in (tqdm(liste_url)):
  n = n+1
  image_data = requests.get(url, headers=headers[f"user{random.randint(1,3)}"]).content
  with open(f'../../Lujes/tables/{n}.jpg', 'wb') as image_file:
      image_file.write(image_data)
  time.sleep(random.randint(3,6))
```
